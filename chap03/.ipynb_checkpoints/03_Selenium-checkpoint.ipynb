```python
elem_login = driver.find_element_by_id("id")
elem_login.clear() # 기존 입력값 지우기
elem_login.send_keys('feb29dy')

elem_login = driver.find_element_by_id("pw")
elem_login.clear()
elem_login.send_keys('password')
```

```python
#Xpath -> javascrpit를 활용하기 위한 웹페이지 내용을 xml 형식으로 보냄
#javascript가 걸려있으면 반드시 xpath 사용
xpath = """//*[@id="frmNIDLogin"]/fieldset/span/input"""
driver.find_element_by_xpath(xpath).click()
```
